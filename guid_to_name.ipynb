```python
def get_unique_guids(tablename, fieldname):
    sql = f"select distinct `{fieldname}` from `{tablename}`;"
    cur.execute(sql)
    res = cur.fetchall()
    return res
```

```python
def get_field_names(tablename):
    sql = f'''SELECT `COLUMN_NAME` 
    FROM `INFORMATION_SCHEMA`.`COLUMNS` 
    WHERE `TABLE_SCHEMA`='esd_phs' 
    AND `TABLE_NAME`='{tablename}';
    '''
    cur.execute(sql)
    fields = cur.fetchall()
    field_names = list()
    for f in fields:
        field_names.append(f[0])

    return field_names
```

```python
def create_field(tablename, columnname):
    sql = f"ALTER TABLE `{tablename}` ADD COLUMN `{columnname}` varchar(60) DEFAULT NULL;"
    cur.execute(sql)
```
